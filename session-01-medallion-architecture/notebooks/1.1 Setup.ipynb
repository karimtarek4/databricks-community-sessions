# 1.1 Setup

Creates the Unity Catalog namespace and managed Volume that all subsequent notebooks depend on.

**What this notebook covers:**
- Creating a catalog and schemas (`landing`, `bronze`, `silver`, `gold`)
- Creating a managed Volume to hold raw data files
- Unity Catalog namespace hierarchy
- Databricks magic commands overview

## Unity Catalog Namespace

| Level       | Also Known As      | Description                                             |
|-------------|--------------------|---------------------------------------------------------|
| `catalog`   | Database           | Top-level container. Groups schemas by domain or env.   |
| `schema`    | Database/Namespace | Logical grouping of related tables within a catalog.    |
| `table`     | Table / View       | The actual data object (managed, external, or view).    |

```sql
CREATE CATALOG IF NOT EXISTS de_lab
COMMENT 'Databricks ETL Masterclass - Session 1';
```

```sql
USE CATALOG de_lab;
SELECT CURRENT_CATALOG();
```

```sql
CREATE SCHEMA IF NOT EXISTS de_lab.landing
  COMMENT 'Landing zone — unmaterialized views over raw Volume files';

CREATE SCHEMA IF NOT EXISTS bronze
  COMMENT 'Bronze layer — deduplicated, type-cleaned Delta tables';

CREATE SCHEMA IF NOT EXISTS silver
  COMMENT 'Silver layer — transformed, business-typed managed tables';

CREATE SCHEMA IF NOT EXISTS gold
  COMMENT 'Gold layer — aggregated, analytics-ready tables';
```

```sql
SHOW SCHEMAS IN de_lab;
```

```sql
CREATE VOLUME IF NOT EXISTS de_lab.landing.de_lab_vol
COMMENT 'Managed volume — holds landing files (CSV, JSON)';
```

```sql
SHOW VOLUMES IN de_lab.landing;
```

## Databricks Magic Commands

| Magic     | Description                                  |
|-----------|----------------------------------------------|
| `%sql`    | Run SQL in a Python-default notebook         |
| `%python` | Run Python in a SQL-default notebook         |
| `%md`     | Render Markdown (documentation)              |
| `%sh`     | Run shell commands on the driver node        |
| `%fs`     | Browse the Databricks file system (DBFS/Volumes) |
